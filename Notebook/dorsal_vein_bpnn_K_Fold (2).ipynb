```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, KFold
from sklearn.preprocessing import LabelEncoder, StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import to_categorical
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report, precision_recall_fscore_support
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the dataset
df = pd.read_csv('identifikasi_dorsalhandvein.csv', header=None, names=['Feature1', 'Feature2', 'Feature3', 'Class'])

# Preprocess the dataset
scaler = StandardScaler()
X = df.drop(['Class'], axis=1)
X = scaler.fit_transform(X)

# Encode labels
y = df['Class']
encoder = LabelEncoder()
y = encoder.fit_transform(y)

# Convert labels to one-hot encoding
y = to_categorical(y)

# Initialize K-Fold Cross-Validation
k_folds = 5
kf = KFold(n_splits=k_folds, shuffle=True, random_state=42)

# Lists to store evaluation metrics
fold_train_accuracies = []
fold_test_accuracies = []
precision_list = []
recall_list = []
f1_score_list = []

# Cross-validation loop
for fold, (train_index, test_index) in enumerate(kf.split(X)):
    print(f"Processing Fold {fold + 1}/{k_folds}")

    # Split data into train and test sets for the current fold
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]

    # Initialize a new model for each fold
    model = Sequential()
    model.add(Dense(8, input_shape=(X_train.shape[1],), activation='relu'))
    model.add(Dense(8, activation='relu'))
    model.add(Dense(y_train.shape[1], activation='softmax'))

    # Compile the model
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

    # Train the model on the current fold
    model.fit(X_train, y_train, epochs=150, batch_size=5, verbose=1)

    # Evaluate on training data
    y_train_pred = np.argmax(model.predict(X_train), axis=1)
    y_train_true = np.argmax(y_train, axis=1)
    train_accuracy = accuracy_score(y_train_true, y_train_pred)
    fold_train_accuracies.append(train_accuracy)

    # Evaluate on testing data
    y_test_pred = np.argmax(model.predict(X_test), axis=1)
    y_test_true = np.argmax(y_test, axis=1)
    test_accuracy = accuracy_score(y_test_true, y_test_pred)
    fold_test_accuracies.append(test_accuracy)

    # Print accuracy for the current fold
    print(f"Fold {fold + 1} Training Accuracy: {train_accuracy * 100:.2f}%")
    print(f"Fold {fold + 1} Testing Accuracy: {test_accuracy * 100:.2f}%")

    # Print classification report for the current fold
    print(f"\nClassification Report for Fold {fold + 1} - Training Data:")
    print(classification_report(y_train_true, y_train_pred, target_names=encoder.classes_))

    print(f"\nClassification Report for Fold {fold + 1} - Testing Data:")
    print(classification_report(y_test_true, y_test_pred, target_names=encoder.classes_))

    # Calculate precision, recall, and f1-score for the current fold
    precision, recall, f1, _ = precision_recall_fscore_support(y_test_true, y_test_pred, average='weighted')
    precision_list.append(precision)
    recall_list.append(recall)
    f1_score_list.append(f1)

    # Confusion Matrix
    cm = confusion_matrix(y_test_true, y_test_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=encoder.classes_, yticklabels=encoder.classes_)
    plt.title(f'Confusion Matrix - Fold {fold + 1}')
    plt.xlabel('Predicted Labels')
    plt.ylabel('True Labels')
    plt.show()

# Calculate and print average metrics across all folds
print(f"\nAverage Training Accuracy across {k_folds} folds: {np.mean(fold_train_accuracies) * 100:.2f}%")
print(f"Average Testing Accuracy across {k_folds} folds: {np.mean(fold_test_accuracies) * 100:.2f}%")
print(f"Average Precision: {np.mean(precision_list) * 100:.2f}%")
print(f"Average Recall: {np.mean(recall_list) * 100:.2f}%")
print(f"Average F1-Score: {np.mean(f1_score_list) * 100:.2f}%")
```
